# Restaurant Sales — Exploratory Data Analysis

This notebook walks through the dataset step by step:
1. Load and inspect the data
2. Top-selling menu items
3. Busiest hours and days
4. Monthly revenue trend
5. Category breakdown
6. Server performance

```python
import sys
sys.path.insert(0, '../scripts')

import pandas as pd
import numpy as np
import plotly.express as px
from clean import load_and_clean
from analyse import (
    top_items_by_revenue, top_items_by_quantity,
    sales_by_hour, sales_by_day,
    monthly_revenue, revenue_by_category,
    server_performance, payment_split
)
```

```python
df = load_and_clean()
print(f"Shape: {df.shape}")
df.head()
```

## Top-selling items

```python
fig = px.bar(
    top_items_by_revenue(df),
    x='revenue', y='menu_item', orientation='h',
    title='Top items by revenue'
)
fig.show()
```

## Busiest hours

```python
fig = px.bar(sales_by_hour(df), x='hour', y='orders', title='Orders by hour')
fig.show()
```

## Revenue by day of week

```python
fig = px.bar(sales_by_day(df), x='day_of_week', y='revenue', title='Revenue by day')
fig.show()
```

## Monthly revenue trend

```python
fig = px.line(monthly_revenue(df), x='month', y='revenue', markers=True, title='Monthly revenue')
fig.show()
```

## Category breakdown

```python
fig = px.pie(revenue_by_category(df), values='revenue', names='category', hole=0.4, title='Revenue by category')
fig.show()
```

## Server performance

```python
print(server_performance(df).to_string(index=False))
```
